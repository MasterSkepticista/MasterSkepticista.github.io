```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Define the data
time_taken = [
  9.1, 
  7.7, 
  6.96, 
  6.61, 
  6.25
]
optimization_methods = [
  '4 GPU 10G baseline',
  '4 GPU 10G baseline\n + gradient compression',
  '4 GPU 10G baseline\n + gradient compression\n + gradient accumulation',
  '4 GPU 10G baseline\n + gradient compression\n + gradient accumulation\n + multi-NIC ',
  'theoretical limit'
]

# Sort the data in descending order
sorted_data = zip(time_taken, optimization_methods)

# Unzip the sorted data
time_taken, optimization_methods = zip(*sorted_data)

# Create the horizontal bar plot
p = plt.barh(optimization_methods[::-1], time_taken[::-1], color='lightblue', align="center", height=0.5)
plt.bar_label(p)

# Color the bar for '4 GPUs theoretical limit' as green
ax = plt.gca()
ax.get_children()[4].set_color('yellow')
ax.get_children()[0].set_color('lightgreen')
ax.get_children()[1].set_color('#D8BFD8')
ax.set_xlim([6, 10])
ax.set_xticks([])
print(ax.get_ylim())

# Set the labels and title
plt.xlabel('ImageNet Training Time (hours)')

# Show the plot
plt.show()
```

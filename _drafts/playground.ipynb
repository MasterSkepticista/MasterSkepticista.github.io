```python
import matplotlib.pyplot as plt

# Data for the plot
categories = ['baseline', 'opt1 (gradient compression)', 'opt2 (gradient accum)', 'opt3 (multi-nic)', 'theoretical limit']
steps_per_sec = [68.8, 81.5, 89.8, 94.5, 100.0]
colors = ['#79c36a', '#34b8db', '#f9e79f', '#f7dc6f', '#f4d03f', '#f39c12', '#e67e22', '#d35400']

# Create the bar plot
plt.figure(figsize=(8, 4))
bars = plt.barh(list(range(len(categories)))[:6], steps_per_sec[:6], color=colors)

# Annotate the bars with the values
for bar, value in zip(bars, steps_per_sec):
    plt.text(bar.get_width() + 0.05, bar.get_y() + bar.get_height()/2, f'{value}', va='center', ha='left', fontsize=10)

# Add labels and title
plt.xlabel('ImageNet Training Time (hours)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks([])  # Remove the y-axis labels
plt.grid(False)
plt.xlim(25, 100)

# Create the legend
plt.legend(bars, categories, loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)

# Adjust layout to mimic the original image
plt.tight_layout()

# Display the plot
plt.show()
```
